```python
import spacy
from spacy import displacy

# -- Beispiel CVs, die nicht zum Trainings verwendet wurden --
CVHebisch = './CVs/EvaluationCVs/notTrainedWith/Allgeier_Hebisch_Torsten.txt' 
CVBrand = './CVs/EvaluationCVs/notTrainedWith/Allgeier_Brand_Arne.txt'
CVSascha = './CVs/EvaluationCVs/notTrainedWith/Allgeier_Hanf_Sascha.txt'
CVMike = './CVs/EvaluationCVs/notTrainedWith/emineo_Götz_Mike.txt'

# -- Modelle die auf den jeweiligen Datensätzen trainiert wurden --
modell_1   = "1"
modell_1_p = "1_Pretrained"
modell_2   = "2"
modell_3_0 = "3_0"
modell_3_1 = "3_1"
modell_3_5 = "3_5"
modell_3_6 = "3_6"
modell_3_7 = "3_7"
modell_4_0 = "4_0"
modell_4_1 = "4_1"
modell_4_2 = "4_2"
modell_4_5 = "4_5"
modell_4_6 = "4_6"

# ----------------------------------
# ------------ Options: ------------
# ----------------------------------

# Einer der oben gelisteten CVs. Auf ihn wird das Modell angewandt
cv     = CVMike
# Eins der oben gelisteten Modelle. Das Modell wird anschließend angewandt
modell = modell_3_0


# ----------------------------------
# ----------- Ausführen: -----------
# ----------------------------------

#CV als txt-Datei einlesen
with open(cv) as f:
    cv = f.read()
       
#Modell importieren
nlp = spacy.load("./Modelle/Modell_{}/model-best/".format(modell))
#CV durch Pipeline laufen lassen
doc = nlp(cv)

#Lädt das dem Modell entsprechende Labelschema
if modell.startswith("1"):
    label = {"Skill:Level:+": "chartreuse", "Skill:Level:0": "DeepSkyBlue", "Skill:Level:-": "orange"}
elif modell.startswith("2"):
    label = {"Skill": "chartreuse"}
elif modell.startswith("3") or modell.startswith("4"):
    label = {"Skill": "DeepSkyBlue", "Tätigkeit": "chartreuse", "Rolle": "LightCoral", "Branche": "MediumPurple"}

#Anschauliche Darstellung der Labels durch Markierung im Text
displacy.render(doc, style="ent", jupyter=True, options={"colors": label})
```
